```python
import pandas as pd
```

```python
dataset=pd.read_csv('car data.csv')
```

```python
dataset.columns
```

```python
dataset=dataset[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
dataset['No_of_year']= 2021 - dataset['Year']
```

```python
dataset.head()
```

```python
dataset.drop(['Year'],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset.head()
```

```python
dataset.corr()
```

```python
import seaborn as sns
```

```python
sns.pairplot(dataset)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline


corrmat=dataset.corr()
top_corr=corrmat.index
plt.figure(figsize=(20,20))
#plotting heat map
g=sns.heatmap(dataset[top_corr].corr(),annot=True,cmap="RdYlGn")
```

```python
#independent and dependent features
#x is all other features than the selling price
#y is the selling price
x=dataset.iloc[:,1:]
y=dataset.iloc[:,0]
```

```python
y.head()
```

```python
x.head()
```

```python
##Feature Importance
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(x,y)
print(model.feature_importances_)
```

```python
#plot graph of each feature for better visualisation
feat_importances = pd.Series(model.feature_importances_,index=x.columns)
#converting the feature importances to a series and returning the top 5
#features hence nlargest(5)
feat_importances.nlargest(5).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
x_train.shape
```

```python
#using a random forest regressor
from sklearn.ensemble import RandomForestRegressor
rf_random=RandomForestRegressor()
```

```python
regressor = RandomForestRegressor()
```

```python
from sklearn.model_selection import RandomizedSearchCV
#RandomizedSearchCV helps us find out the best features and the number of
#features we should have
#This SearchCV is faster than gridsearchCV
#automatically finds out the best parameter
```

```python
##Hyperparameters tuning is done here
import numpy as np

#Number of trees in random forest
n_estimators=[int(x) for x in np.linspace(start=100,stop=1200,num=12)]

#Number of features to consider at every split
max_features=['auto','sqrt']

#Maximum number of levels in the tree
max_depth=[int(x) for x in np.linspace(5, 30, num = 6)]
max_depth.append(None)

#Minimum number of samples required to split a node
min_samples_split = [2,5,10,15,100]

#Minimum number of samples required for each leaf node
min_samples_leaf=[1,2,5,10]
```

```python
#Create the random grid
random_grid = {'n_estimators' : n_estimators,
               'max_features' : max_features,
               'max_depth' : max_depth,
               'min_samples_split' : min_samples_split,
               'min_samples_leaf' : min_samples_leaf}
print(random_grid)
```

```python
#Use random grid to search for best hyperparameters
#create base model to tune
rf=RandomForestRegressor()
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
#cv is cross validation
#verbose is so that we can get the values displayed 
rf_random = RandomizedSearchCV(estimator = rf,param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf.get_params().keys()
```

```python
rf_random.fit(x_train,y_train)
```

```python
rf_random.best_params_
```

```python
predictions=rf_random.predict(x_test)
```

```python
predictions
```

```python
rf_random.best_score_
```

```python
sns.distplot(y_test-predictions)
#normal distribution which means,we have good result
#y test and predictions should be equal to each other so difference should be
#minimum
```

```python
plt.scatter(y_test,predictions)
#linear plotting means prediction is good
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
